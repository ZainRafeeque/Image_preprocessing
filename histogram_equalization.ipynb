### Histogram Equalization

Histogram equalization is a method in image processing of contrast adjustment using the imageâ€™s histogram. It aims to distribute the intensity values of the image more uniformly, so that the image utilizes the full range of possible values from 0 (black) to 255 (white). This can enhance the contrast and details of the image.

```python
img = cv2.imread('albert-einstein.jpg',0) # load image in grayscale mode
equ = cv2.equalizeHist(img) # apply histogram equalization
cv2.imwrite('equalized_image.jpg',equ) # save the result

# plot the original and equalized histograms
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.hist(equ.flatten(),256,[0,256], color = 'b')
plt.xlim([0,256])
plt.legend(('Original','Equalized'), loc = 'upper left')
plt.show()
```
