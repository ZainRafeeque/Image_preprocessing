```python
from skimage import morphology
from skimage import io
from matplotlib import pyplot as plt
```

#### Morphological operations 
are image processing techniques that modify the shape and structure of objects in an image. They are often used for noise removal, edge detection, segmentation, etc. Some common morphological operations are:

```python
img = io.imread('albert-einstein.jpg', as_gray=True) # load image in grayscale mode
selem = morphology.disk(6) # create a disk-shaped structuring element
```

#### Erosion
is one of the basic morphological operations in image processing. It is used to reduce the size of the foreground objects in an image by removing pixels from their boundaries. It can also be used to remove noise, isolate individual elements, or find holes in an image.

Erosion works by sliding a small shape or kernel over the image and replacing each pixel by the minimum value of its neighboring pixels that are covered by the kernel. The shape and size of the kernel affect the result of the erosion operation. A larger kernel will erode more pixels than a smaller one



```python
eroded = morphology.erosion(img, selem) # apply erosion
# plot the original and processed images
fig, axes = plt.subplots(1, 2, figsize=(6, 6))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(eroded, cmap=plt.cm.gray)
ax[1].set_title('Erosion')

for a in ax:
    a.axis('off')
plt.tight_layout()
plt.show()
```

#### Dilation 
is the opposite of erosion in morphological operations. It is used to increase the size of the foreground objects in an image by adding pixels to their boundaries. It can also be used to fill gaps, connect disjoint elements, or thicken lines in an image.

Dilation works by sliding a small shape or kernel over the image and replacing each pixel by the maximum value of its neighboring pixels that are covered by the kernel. The shape and size of the kernel affect the result of the dilation operation. A larger kernel will dilate more pixels than a smaller one

```python
eroded = morphology.erosion(img, selem) # apply erosion
dilated = morphology.dilation(img, selem) # apply dilation
opened = morphology.opening(img, selem) # apply opening
closed = morphology.closing(img, selem) # apply closing

# plot the original and processed images
fig, axes = plt.subplots(1, 3, figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(eroded, cmap=plt.cm.gray)
ax[1].set_title('Erosion')

ax[2].imshow(dilated, cmap=plt.cm.gray)
ax[2].set_title('Dilation')

for a in ax:
    a.axis('off')
plt.tight_layout()
plt.show()
```

#### Opening 
is a morphological operation that combines erosion and dilation. It is used to remove small objects or noise from an image while preserving the shape and size of larger objects. It can also be used to smooth the contours, separate disjoint objects, or eliminate thin bridges or gaps in an image.

Opening works by first applying erosion to an image with a structuring element, and then applying dilation to the eroded image with the same structuring element. The structuring element determines the size and shape of the objects that are removed or retained by the opening operation

```python
selem = morphology.disk(6) # create a disk-shaped structuring element
eroded = morphology.erosion(img, selem) # apply erosion
dilated = morphology.dilation(img, selem) # apply dilation
opened = morphology.opening(img, selem) # apply opening
closed = morphology.closing(img, selem) # apply closing

# plot the original and processed images
fig, axes = plt.subplots(2, 3, figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(eroded, cmap=plt.cm.gray)
ax[1].set_title('Erosion')

ax[2].imshow(dilated, cmap=plt.cm.gray)
ax[2].set_title('Dilation')

ax[3].imshow(opened, cmap=plt.cm.gray)
ax[3].set_title('Opening')


for a in ax:
    a.axis('off')
plt.tight_layout()
plt.show()
```

#### Closing 
is a morphological operation that combines dilation and erosion. It is used to fill small holes or gaps in the foreground objects in an image while preserving the shape and size of the objects. It can also be used to smooth the contours, connect disjoint objects, or eliminate thin breaks or cracks in an image.

Closing works by first applying dilation to an image with a structuring element, and then applying erosion to the dilated image with the same structuring element. The structuring element determines the size and shape of the holes or gaps that are filled or retained by the closing operation1

```python
eroded = morphology.erosion(img, selem) # apply erosion
dilated = morphology.dilation(img, selem) # apply dilation
opened = morphology.opening(img, selem) # apply opening
closed = morphology.closing(img, selem) # apply closing

# plot the original and processed images
fig, axes = plt.subplots(2, 3, figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(eroded, cmap=plt.cm.gray)
ax[1].set_title('Erosion')

ax[2].imshow(dilated, cmap=plt.cm.gray)
ax[2].set_title('Dilation')

ax[3].imshow(opened, cmap=plt.cm.gray)
ax[3].set_title('Opening')

ax[4].imshow(closed, cmap=plt.cm.gray)
ax[4].set_title('Closing')

for a in ax:
    a.axis('off')
plt.tight_layout()
plt.show()
```

#### Edge Detection
Edge detection is a technique to find the boundaries of objects or regions in an image. Morphological operations are methods to manipulate the shape and structure of objects in an image using a structuring element. Morphological operations can be used for edge detection by combining erosion, dilation, opening, closing, or other operations to enhance or extract the edges.

One common way to perform edge detection using morphological operations is to use the morphological gradient, which is the difference between the dilation and erosion of an image with the same structuring element. The morphological gradient highlights the edges of the objects in the image

```python
# Import OpenCV library
import cv2

# Read an input image
image = cv2.imread("albert-einstein.jpg")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detector with threshold values 30 and 100
edged_image = cv2.Canny(gray_image, 30, 100)

# Save the output image
cv2.imwrite("edge_detection.jpg", edged_image)
plt.imshow(edged_image, cmap='gray')
```

#### Image segmentation 
is the process of dividing an image into meaningful regions or segments based on some criteria. It can be useful for object detection, recognition, analysis, etc. There are different ways to perform image segmentation in OpenCV Python, such as:

K-Means: Clusters the pixels based on their color or intensity values. Assigns each pixel to one of the K clusters.
Contour Detection: Finds the boundaries of the objects in an image. Draws curves around the regions with similar pixels.
Thresholding: Converts a grayscale image into a binary image by setting a threshold value. Assigns 0 or 1 to each pixel depending on whether it is below or above the threshold.
Color Masking: Filters out the pixels that do not match a specific color range. Creates a mask that only shows the pixels within the desired color.

**Example** of image segmentation using thresholding technique to segment the image into foreground and background.

```python
import cv2

# Read the image
img = cv2.imread('albert-einstein.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply a threshold to the grayscale image
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Start a window thread
cv2.startWindowThread()

# Display the original and segmented images
cv2.imshow('Original', img)
cv2.imshow('Segmented', thresh)

# Wait for a key press or a time limit to close the window
cv2.waitKey(0) & 0xFF==ord('q')

# Destroy the window
cv2.destroyAllWindows()

# Wait for another key press or a time limit to close the window
cv2.waitKey(1)
# plot the original and processed images
fig, axes = plt.subplots(1, 2, figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(thresh, cmap=plt.cm.gray)
ax[1].set_title('Segmented')

for a in ax:
    a.axis('off')
plt.tight_layout()
plt.show()
```
